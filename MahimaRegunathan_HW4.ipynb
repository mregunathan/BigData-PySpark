***Task 12. Calculate the count of rows where the gt column contains stairs only for users a and b on Device=nexus4_1***

```
import pyspark.sql.functions as F
stairsonlyDF = streamingDF.where((F.col("gt").contains("stairs")) & ((F.col("User") == "a") | (F.col("User") == "b")) & (F.col("Device") == "nexus4_1"))
stairsonlyDF = stairsonlyDF.select("User").groupby("User").count()
```

```
stairsonlyQuery = stairsonlyDF.writeStream\
  .outputMode("complete")\
  .format("memory")\
  .queryName("stairsonlydata")\
  .start()
```

```

for x in range(5):
  spark.sql("SELECT * FROM stairsonlydata").show()
  sleep(5)
```

```
stairsonlyQuery.stop()
```

***Task 13. Compute the average of colummns x, y, x for each combination of values***

```
gtmodelstatsDF = streamingDF.groupby(["gt", "model"]).agg(F.avg("x").alias("x_avg"), F.avg("y").alias("y_avg"), F.avg("z").alias("z_avg"), F.count("Device").alias("devicecount"))
```

```
gtmodelstatsQuery = gtmodelstatsDF.writeStream\
  .format("memory")\
  .outputMode("complete")\
  .queryName("gtmodelstatsdata")\
  .start()
```

```
from time import sleep
for x in range(5):
  spark.sql("SELECT * FROM gtmodelstatsdata").where(F.col("x_avg") > 0).show()
  sleep(5)
```

```
gtmodelstatsQuery.stop()
```

***Task 14. Perform a join of a static DataFrame with a streaming DataFrame***

```
historicalAvg = staticDF.groupby(["gt", "model"]).agg(F.avg("x").alias("x_avg1"), F.avg("y").alias("y_avg1"), F.avg("z").alias("z_avg1"))
historicalAvg.show()
```

```
streamingAvg = streamingDF.groupby(["gt", "model"]).agg(F.avg("x").alias("x_avg2"), F.avg("y").alias("y_avg2"), F.avg("z").alias("z_avg2"))
```

```
jointransformDF = historicalAvg.join(streamingAvg, on = ["gt", "model"])
```

```
joinQuery = jointransformDF.writeStream\
  .format("memory")\
  .outputMode("complete")\
  .queryName("jointransformdata")\
  .start()
```

```
for x in range(5):
  spark.sql("SELECT * FROM jointransformdata").show()
  sleep(5)
```

```
joinQuery.stop()
```
